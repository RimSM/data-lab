```python
import pyspark 
```

```python
sc = pyspark.SparkContext.getOrCreate()
```

```python
rdd = sc.parallelize([("name", "joe,sarah,tom"), ("car", "hyundai")])
result1 = rdd.map(lambda x: x[1].split(','))
print(result1.collect())
```

```python
#rdd = sc.parallelize([("name", "joe,sarah,tom"), ("car", "hyundai")])
result2 = rdd.flatMap(lambda x: x[1].split(','))
print(result2.collect())
```

```python
test_file = "file:///home/jovyan/work/sample/lorem_ipsum.txt"
lines = sc.textFile(test_file)
#print(lines.collect())

words = lines.flatMap(lambda x: x.split())
#print(words.collect())

word_count = words.countByValue()
for word, count in word_count.items():
    print(f"{word}: {count}")
```

```python
word_count = words.map(lambda x: (x,1)).reduceByKey(lambda x,y: x+y)
sorted_word_count = word_count.map(lambda x: (x[1], x[0])).sortByKey(False)

#print(word_count.collect())
#print(sorted_word_count.collect())
for count, word in sorted_word_count.collect() : 
    print(f'{count}: {word}')
```
